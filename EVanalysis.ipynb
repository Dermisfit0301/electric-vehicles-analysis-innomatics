**Electrical Vehicle EDA**


 We are going to get several insights from an excel dataset about electrical vehicles in the USA. We are going to do exploratory data analysis and use plotly to visualize the data. The data cleaning and normalization is going to be skipped as our main focus would be on exploring the dataset rather than creating a model.

```python
#getting packages
import pandas as pd
import numpy as np
import plotly.express as px
import bar_chart_race as bcr
```

```python
#getting the dataset

ev=pd.read_csv("/content/electriccardataset.csv")
ev.head()
```

```python
#examining the dataset
ev.info()
```

```python
#examining the shape
ev.shape
```

```python
# column name cleaning
ev.columns = ev.columns.str.replace(' ', '')
ev.columns
```

```python
#checking null values
ev.isna().sum(axis=0)
```

```python
#dropping null values
ev = ev.dropna()
ev
```

```python
# Calculating the count of each Electric Vehicle Type
electric_vehicle_count = ev['ElectricVehicleType'].value_counts()

# percentage of each Electric Vehicle Type
electric_vehicle_percentages = (electric_vehicle_count / electric_vehicle_count.sum()) * 100

# creating pie chart
fig = px.pie(names=electric_vehicle_count.index, values=electric_vehicle_percentages,
             title='Percentage of Electric Vehicle Types')
fig.show()
```

We can infer from the pie chart that 3 in 4 electric vehicles are plug-in hybrids.

```python
# number of ids per make type
make_dol_count = ev.groupby('Make')['DOLVehicleID'].nunique().reset_index()

# bar chart
fig = px.bar(make_dol_count, x='Make', y='DOLVehicleID',
             title='Number of DOL Vehicle IDs for Each Make',
             labels={'Make': 'Make', 'DOL Vehicle ID': 'Number of DOL Vehicle IDs'})
fig.show()
```

We can see that Elon Musk's Tesla has the highest number of DOL Vehicle IDs. Hence we can see that Tesla is the most popular electric vehicle brand in the USA.

```python
# extraction of longitude and latitude
ev[['Longitude', 'Latitude']] = ev['VehicleLocation'].str.extract(r'\((-?\d+\.\d+)\s(-?\d+\.\d+)\)', expand=True).astype(float)

# chloropleth
fig = px.scatter_mapbox(ev,
                        lat='Latitude',
                        lon='Longitude',
                        size='DOLVehicleID',
                        size_max=20,
                        hover_name='DOLVehicleID',
                        hover_data={'Latitude': False, 'Longitude': False},
                        mapbox_style='open-street-map',
                        title='Scatter Plot of Number of DOL IDs at Different Locations')

fig.update_layout(showlegend=False)

fig.show()
```

```python
#  Box Plot of cafv eligibility and electric range
fig = px.box(ev, x='CleanAlternativeFuelVehicle(CAFV)Eligibility', y='ElectricRange',
             title='Box Plot of Electric Range by CAFV Eligibility',
             labels={'Clean Alternative Fuel Vehicle (CAFV) Eligibility': 'CAFV Eligibility', 'Electric Range': 'Electric Range (in miles)'})

# Show the plot
fig.show()
```

We can infer almost all the vehicles are eligible and hence create little to no pollution

```python

# Create the cross-tabulation between model and electric utility
cross_tab = pd.crosstab(ev['Model'], ev['ElectricUtility'])

# Display the cross-tabulation
print(cross_tab)
```
