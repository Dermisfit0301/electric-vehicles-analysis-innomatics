**Electrical Vehicle EDA**


 We are going to get several insights from an excel dataset about electrical vehicles in the USA. We are going to do exploratory data analysis and use plotly to visualize the data. The data cleaning and normalization is going to be skipped as our main focus would be on exploring the dataset rather than creating a model.

```python
#getting packages
import pandas as pd
import numpy as np
import plotly.express as px
import bar_chart_race as bcr
```

```python
#getting the dataset

ev=pd.read_csv("/content/electriccardataset.csv")
ev.head()
```

```python
#examining the dataset
ev.info()
```

```python
#examining the shape
ev.shape
```

```python
# column name cleaning
ev.columns = ev.columns.str.replace(' ', '')
ev.columns
```

```python
#checking null values
ev.isna().sum(axis=0)
```

```python
#dropping null values
ev = ev.dropna()
ev
```

```python
# Calculating the count of each Electric Vehicle Type
electric_vehicle_count = ev['ElectricVehicleType'].value_counts()

# percentage of each Electric Vehicle Type
electric_vehicle_percentages = (electric_vehicle_count / electric_vehicle_count.sum()) * 100

# creating pie chart
fig = px.pie(names=electric_vehicle_count.index, values=electric_vehicle_percentages,
             title='Percentage of Electric Vehicle Types')
fig.show()
```

We can infer from the pie chart that 3 in 4 electric vehicles are plug-in hybrids.

```python
#electric range distribution
data = ev['ElectricRange']

# Create a histogram
fig = px.histogram(data, nbins=10, title='Histogram of Electric Range Distribution')
fig.update_layout(xaxis_title='Electric Range', yaxis_title='Frequency')
fig.show()
```

We see most vehicles are less than 100 in electric range units.

```python
# Let's assume 'City' and 'DOL Vehicle ID' are the columns in your DataFrame 'df'
data = ev.groupby('City')['DOLVehicleID'].count().reset_index()
data = data.sort_values(by='DOLVehicleID', ascending=False).head(5)

# Create a bar chart
fig = px.bar(data, x='City', y='DOLVehicleID', title='Top 5 Cities with Highest number of Electric vehicles')
fig.show()
```

Seattle has the most number of electric vehicles in the country as of now.

```python
# number of ids per make type
make_dol_count = ev.groupby('Make')['DOLVehicleID'].nunique().reset_index()

# bar chart
fig = px.bar(make_dol_count, x='Make', y='DOLVehicleID',
             title='Number of DOL Vehicle IDs for Each Make',
             labels={'Make': 'Make', 'DOL Vehicle ID': 'Number of DOL Vehicle IDs'})
fig.show()
```

We can see that Elon Musk's Tesla has the highest number of DOL Vehicle IDs. Hence we can see that Tesla is the most popular electric vehicle brand in the USA.

```python
#  Box Plot of cafv eligibility and electric range
fig = px.box(ev, x='CleanAlternativeFuelVehicle(CAFV)Eligibility', y='ElectricRange',
             title='Box Plot of Electric Range by CAFV Eligibility',
             labels={'Clean Alternative Fuel Vehicle (CAFV) Eligibility': 'CAFV Eligibility', 'Electric Range': 'Electric Range (in miles)'})

# Show the plot
fig.show()
```

We can say that the vehicles that are eligible for clean alternative fuel are the ones having the highest electric range.

```python
# Group the data by 'Make' and 'Model Year' and calculate the count of electric vehicles
grouped_data = ev.groupby(['Make', 'ModelYear']).size().reset_index(name='Count')

# Create the grouped bar plot
fig = px.bar(grouped_data,
             x='Make',
             y='Count',
             color='ModelYear',
             title='Count of Electric Vehicles by Make and Model Year',
             labels={'Make': 'Make', 'Count': 'Count of Electric Vehicles'})

fig.show()
```

We can say that most of the electic vehicles are made by tesla and in general the period of 2015 to 2020 saw the most number of electric vehicles being manufactured.

```python
#Grouping the data
df_count = ev.groupby('ModelYear')['DOLVehicleID'].count().reset_index()

#  time series graph
fig = px.line(df_count,
              x='ModelYear',
              y='DOLVehicleID',
              title='Time Series of Count of DOL Vehicle IDs by Model Year',
              labels={'Model Year': 'ModelYear', 'DOLVehicleID': 'Count of DOL Vehicle IDs'})

fig.show()
```

Clearly we can see the early decade was very stagnant and almost no electric vehicle was manufactured. After 2010 the manufacturing of electric vehicle spiked and it peaked a bit in 2015 and reached the top at 2020 where 5000 electric vehicles were created and after 2020 there was a sharp dip.

```python
# Group by 'Make' and 'Model' and count the occurrences of each model for each make
popular_model_df = ev.groupby(['Make', 'Model']).size().reset_index(name='Count')

# Find the popular model for each make (model with the highest count)
popular_model_df = popular_model_df.sort_values(by='Count', ascending=False).drop_duplicates(subset='Make', keep='first')

# Create a bar plot to visualize the popular model for each make
fig = px.bar(popular_model_df, x='Make', y='Count', text='Model', title='Popular Model for Each Make')
fig.update_traces(textposition='inside')
fig.show()
```

We can see that Tesla, Nissan and Chevrolet are the most popular brands for electrical vehicles. Model 3 is the most selling Tesla EV whil leaf is most selling Nissan model and BOLTEV is the most selling chevrolet model.

```python

# Create the cross-tabulation between model and electric utility
cross_tab = pd.crosstab(ev['Model'], ev['ElectricUtility'])

# Convert the 'data' dictionary to a DataFrame
cross_tab = pd.DataFrame(data=cross_tab)
cross_tab
```

```python
# Extracting longitude and latitude from 'VehicleLocation'
ev[['Longitude', 'Latitude']] = ev['VehicleLocation'].str.extract(r'\((-?\d+\.\d+)\s(-?\d+\.\d+)\)', expand=True).astype(float)

# Group by location and calculate the count of DOLVehicleID
location_count = ev.groupby(['Longitude', 'Latitude'])['DOLVehicleID'].count().reset_index(name='Count')

# Scatter map with Choropleth effect showing vehicle id counts
fig = px.scatter_mapbox(location_count,
                        lat='Latitude',
                        lon='Longitude',
                        size='Count',
                        size_max=20,
                        hover_name='Count',
                        hover_data={'Longitude': False, 'Latitude': False},
                        mapbox_style='open-street-map',
                        title='Choropleth Map of Number of DOL IDs at Different Locations')

fig.update_layout(showlegend=False)

fig.show()
```

```python
# subsetting the dataset
data = {
    'ModelYear': ev['ModelYear'],
    'Make': ev['Make']
}
# Convert the 'data' dictionary to a DataFrame
df = pd.DataFrame(data=data)
df
```

```python


# Grouping the data  to get counts for each make type in each year
df_grouped = df.groupby(['ModelYear', 'Make']).size().reset_index(name='Count')

# Pivot the DataFrame to get counts of each make type by year
df_count = df_grouped.pivot_table(index='ModelYear', columns='Make', values='Count', fill_value=0)

# Creation of running bar chart
bcr.bar_chart_race(
    df=df_count,
    filename="ev_make_count_animation.mp4",
    orientation="h",
    sort="desc",
    n_bars=10,
    steps_per_period=45,
    period_length=3000,
    title={'label': 'Year wise count of each Make Type',
           'size': 52,
           'weight': 'bold',
           'pad': 40
           },
    period_label={'x': .95, 'y': .15,
                  'ha': 'right',
                  'va': 'center',
                  'size': 72,
                  'weight': 'semibold'
                  },


    bar_kwargs={'alpha': .99, 'lw': 0},

)
```
